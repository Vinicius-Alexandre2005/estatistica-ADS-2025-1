```python
# Exercício 1:

import statistics

cargas = [5, 4, 8, 8, 7, 6, 6, 8, 8, 12]

# Média:
media = sum(cargas) / len(cargas)

# Mediana:

mediana = statistics.median(cargas)

# Moda:
moda = statistics.mode(cargas)



print(f"A média da carga horária dos vendedores é: {media:.2f} horas")
print(f"A mediana da carga horária dos vendedores é: {mediana} horas")
print(f"A moda da carga horária dos vendedores é: {moda} horas")
```

```python
# Exercício 3:

import statistics

valores = [
  120, 250, 250, 251, 251, 785, 458, 124, 245, 125,

  201, 365, 654, 568, 251, 365, 145, 896, 145, 256,

  896, 785, 456, 256, 123, 154, 452, 258, 145, 695,

  145, 254, 654, 563, 562, 456, 125, 145, 258, 145,

  145, 895, 145, 785, 458, 800, 900, 400, 500, 365,

  254, 365, 154, 215, 458, 254, 258, 145, 369, 547,

  589, 698, 698, 789, 544, 456, 356, 548, 569, 598
]

# Média moda e mediana:

# Média:
media = sum(valores)/len(valores)

# Mediana:
mediana = statistics.median(valores)

# Moda:
moda = statistics.mode(valores)

print(f"A média dos valores é: R${media:.2f}")
print(f"A mediana dos valores é: R${mediana}")
print(f"A moda dos valores é: R${moda}")

# Tabela:

num_classes = 8

min_val = min(valores)
max_val = max(valores)
range_val = max_val - min_val

intervalo = np.ceil(range_val / num_classes)

classes = np.arange(min_val, max_val + intervalo, intervalo)

frequências, bins = np.histogram(valores, bins=classes)

tabela = pd.DataFrame({
    'Classe (Intervalo)': [f'{int(bins[i])} - {int(bins[i+1])}' for i in range(len(bins)-1)],
    'Frequência': frequências,
    'Frequência Acumulada': np.cumsum(frequências)
})

print(tabela)
```

```python
# Exercício 4:

import statistics
import pandas as pd
import numpy as np

# Moda, média e mediana:

dias = [
  150, 210, 309, 270, 180, 246, 285, 195, 210, 248,

  199, 250, 290, 195, 301, 221, 210, 190, 210, 259
]

# Média:
media = sum(dias)/len(dias)

# Mediana:
mediana = statistics.median(dias)

# Moda:
moda = statistics.mode(dias)

print(f"A média de dias de duração das lâmpadas é de: {media:.2f} dias")
print(f"A mediana de dias de duração das lâmpadas é de: {mediana} dias")
print(f"A moda de dias de duração das lâmpadas é de: {moda} dias")

#Tabela:

num_classes = 6

min_val = min(dias)
max_val = max(dias)
range_val = max_val - min_val

intervalo = np.ceil(range_val / num_classes)

classes = np.arange(min_val, max_val + intervalo, intervalo)

frequências, bins = np.histogram(dias, bins=classes)

tabela = pd.DataFrame({
    'Classe (Intervalo)': [f'{int(bins[i])} - {int(bins[i+1])}' for i in range(len(bins)-1)],
    'Frequência': frequências,
    'Frequência Acumulada': np.cumsum(frequências)
})

print(tabela)
```
