```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
```

```python
abas = pd.read_excel('/content/drive/MyDrive/Regressão_exercícios.xls', sheet_name=None)
```

```python
# Acesso às planilhas do arquivo Excel
df_combustivel = abas['Economia de Combustível']
df_fundos = abas['Fundos']
df_habitacoes = abas['Notas das Casas']
df_consumo = abas['Consumo']
```

# Economia

```python
# Seleção da variável alvo e das preditoras
target_combustivel = df_combustivel['Economia']
features_combustivel = df_combustivel[['Deslocamento ', 'Cilindrada']]

# Inclusão do intercepto no modelo
features_combustivel_const = sm.add_constant(features_combustivel)

# Criação e ajuste do modelo
reg_combustivel = sm.OLS(target_combustivel, features_combustivel_const).fit()

# Exibição dos resultados
print(reg_combustivel.summary())

# Fórmula estimada do modelo
print('Economia = 24.5071 - 3.7488 * Deslocamento + 1.0466 * Cilindrada')
```

# Notas das casas

```python
# Definição da variável dependente e das explicativas
target_casas = df_habitacoes['NotaFinal']
features_casas = df_habitacoes[['NotaDistancia', 'NotaRecursos']]

# Adicionando a constante
features_casas_const = sm.add_constant(features_casas)

# Ajuste do modelo
reg_casas = sm.OLS(target_casas, features_casas_const).fit()

# Sumário do modelo
print(reg_casas.summary())

# Equação do modelo
print('NotaFinal = 33.4848 + 1.8998 * NotaDistancia + 2.6108 * NotaRecursos')
```

# Fundos

```python
# Definindo variáveis do modelo
target_fundos = df_fundos['TaxadeRetorno']
features_fundos = df_fundos[['ClassificacaoSegurança', 'DespesaAnual']]

# Inclusão do termo constante
features_fundos_const = sm.add_constant(features_fundos)

# Treinamento da regressão
reg_fundos = sm.OLS(target_fundos, features_fundos_const).fit()

# Exibição do sumário
print(reg_fundos.summary())

# Fórmula estimada
print('Taxa de Retorno = 247.3579 - 32.8445 * Segurança + 34.5887 * DespesaAnual')
```

# Consumo

```python
# Variáveis para análise
target_imposto = df_consumo['Imposto']
features_imposto = df_consumo[['Renda', 'Tamanho Familia']]

# Constante para o modelo
features_imposto_const = sm.add_constant(features_imposto)

# Modelo OLS
reg_imposto = sm.OLS(target_imposto, features_imposto_const).fit()

# Resultados
print(reg_imposto.summary())

# Expressão da regressão
print('Imposto = 1304.9048 - 33.1330 * Renda + 356.2959 * Tamanho da Família')
```
